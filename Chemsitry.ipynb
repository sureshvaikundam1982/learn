```python


## 1. The Baking Soda Volcano (Chemical Reaction)



### Code for Simulation:

import matplotlib.pyplot as plt
import numpy as np

# Simulate the eruption using a sine wave to represent bubbles of CO2
x = np.linspace(0, 10, 100)
y = np.sin(x) * np.exp(-x/5)  # Damped sine wave to simulate eruption bubbles

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Baking Soda and Vinegar Reaction", color="red")
plt.title("Baking Soda Volcano - Chemical Reaction")
plt.xlabel("Time (seconds)")
plt.ylabel("Bubble Height (arbitrary units)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Simulate crystal growth by plotting points of crystal growth
crystals_x = np.random.uniform(0, 10, 100)
crystals_y = np.random.uniform(0, 10, 100)

plt.figure(figsize=(8, 6))
plt.scatter(crystals_x, crystals_y, c="blue", label="Growing Crystals")
plt.title("Simulated Salt Crystal Growth")
plt.xlabel("X Position (arbitrary units)")
plt.ylabel("Y Position (arbitrary units)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate the eggshell dissolving with a shrinking circle (representing the egg)
time = np.linspace(0, 10, 100)
shell_thickness = np.exp(-time / 2)  # Exponentially decreasing thickness

plt.figure(figsize=(8, 6))
plt.plot(time, shell_thickness, label="Eggshell Thickness", color="green")
plt.title("The Egg in Vinegar - Chemical Reaction")
plt.xlabel("Time (hours)")
plt.ylabel("Shell Thickness (arbitrary units)")
plt.grid(True)
plt.legend()
plt.show()
```
