```python
# 📘 Simple Lab: Analyzing a Linear Relationship

# 1. Import Libraries
import numpy as np
import matplotlib.pyplot as plt

# 2. Create Sample Data
x = np.linspace(0, 10, 50)
y = 2 * x + 1 + np.random.normal(0, 1, 50)  # y = 2x + 1 + noise

# 3. Plot the Data
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='yellow', label='Data points')
plt.title("Scatter Plot of x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# 4. Fit a Linear Model
m, b = np.polyfit(x, y, 1)
y_fit = m * x + b

# 5. Plot Fitted Line
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label='Data')
plt.plot(x, y_fit, color='pink', label=f'Fitted line: y = {m:.2f}x + {b:.2f}')
plt.title("Linear Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# 6. Conclusion
print(f"The line of best fit is: y = {m:.2f}x + {b:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a circle
fig, ax = plt.subplots()

# Set the aspect ratio to be equal, so the circle is not distorted
ax.set_aspect('equal')

# Add a circle: center (0,0), radius 1
circle = plt.Circle((0, 0), 2, edgecolor='b', facecolor='g', linewidth=2)
ax.add_patch(circle)

# Set the limits of the plot
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)

plt.title("Circle")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a rectangle
fig, ax = plt.subplots()

# Add a rectangle: (x, y), width, height
rectangle = plt.Rectangle((0, 0), 2, 1, edgecolor='g', facecolor='yellow', linewidth=2)
ax.add_patch(rectangle)

# Set the limits of the plot
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 2)

plt.title("Rectangle")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a triangle
fig, ax = plt.subplots()

# Triangle coordinates
triangle = plt.Polygon([(0, 0), (2, 0), (1, 2)], edgecolor='purple', facecolor='orange', linewidth=2)
ax.add_patch(triangle)

# Set the limits of the plot
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 3)

plt.title("Triangle")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Create an ellipse
fig, ax = plt.subplots()

# Add an ellipse: center (0, 0), width 3, height 2
ellipse = Ellipse((0, 0), width=3, height=2, edgecolor='r', facecolor='cyan', linewidth=2)
ax.add_patch(ellipse)

# Set the limits of the plot
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

plt.title("Ellipse")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

# Coordinates for a custom polygon
x = [0, 1, 2, 3, 4]
y = [0, 2, 0, 2, 0]

# Create a figure
fig, ax = plt.subplots()

# Plot the polygon using the coordinates
ax.fill(x, y, edgecolor='purple', facecolor='yellow', linewidth=2)

# Set the limits of the plot
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 3)

plt.title("Custom Polygon")
plt.grid(True)
plt.show()
```
